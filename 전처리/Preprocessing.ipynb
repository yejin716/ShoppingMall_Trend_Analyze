```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time

# 1. 드라이버 설정
chrome_options = Options()
chrome_options.add_argument("--headless")  # 브라우저 안 띄우기 (선택)
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# 2. 브랜드가 NaN인 링크 리스트 추출
null_links = df[df['브랜드'].isna()]['링크'].unique().tolist()

# 3. 하나씩 방문해서 브랜드명 가져오기
for link in null_links:
    try:
        driver.get(link)
        time.sleep(2)  # 로딩 대기 (필요시 더 늘리기)

        # 브랜드명 크롤링
        brand_name = driver.find_element(By.CSS_SELECTOR, '.css-gwr30y').text.strip()

        # DataFrame의 해당 링크 행에 브랜드명 삽입
        df.loc[(df['링크'] == link) & (df['브랜드'].isna()), '브랜드'] = brand_name
        print(f"✅ {link} → 브랜드명: {brand_name}")
    
    except Exception as e:
        print(f"⚠️ {link} → 크롤링 실패: {e}")

# 4. 드라이버 종료
driver.quit()
```
